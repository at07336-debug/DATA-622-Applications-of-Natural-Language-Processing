```python
import random
import re
from collections import defaultdict


text = """
The battle between Marvel and DC heroes began in the skies above the city.
Superman flew across the skyline while Iron Man prepared his armor systems.
Batman watched carefully from the shadows as Captain America raised his shield.
Wonder Woman charged forward with her sword while Thor summoned thunder from the clouds.
Spider Man swung between buildings trying to distract the powerful Darkseid.
Doctor Strange opened portals while The Flash moved faster than anyone could see.
Hulk smashed through the streets as Aquaman summoned waves from the ocean.
The heroes realized the battle was not about winning but about protecting the city and its people.
Together the Marvel and DC heroes stood strong against the chaos threatening the world.
"""

# Clean and tokenize words
words = re.findall(r"\b\w+\b|[.,!?;]", text.lower())

# Build Markov chain dictionary
markov_chain = defaultdict(list)

for i in range(len(words) - 1):
    markov_chain[words[i]].append(words[i + 1])

# Starting phrase
start_phrase = "the battle between marvel and dc heroes"
generated_words = start_phrase.lower().split()

current_word = generated_words[-1]

# Generate story
for _ in range(180):
    if current_word in markov_chain:
        next_word = random.choice(markov_chain[current_word])
        generated_words.append(next_word)
        current_word = next_word
    else:
        break

# Convert to readable text
story = " ".join(generated_words)
story = story.replace(" ,", ",").replace(" .", ".").replace(" !", "!").replace(" ?", "?")

print("Generated Markov Chain Story:\n")
print(story)
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = """
Write a creative short story answering the question:
'Why Study Data Science Today?'
Make it sound like a real story, not an essay or bullet points.
"""

result = generator(
    prompt,
    max_length=250,
    num_return_sequences=1,
    truncation=True
)

print("LLM-Generated Story:\n")
print(result[0]["generated_text"])
```
