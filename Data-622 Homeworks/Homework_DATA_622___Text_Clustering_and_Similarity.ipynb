## Exercise 1 — K-Means Clustering

```python
!pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

documents = [
    "The plaintiff filed a lawsuit for breach of contract.",
    "Contract disputes are common in civil litigation.",
    "The defendant denies all allegations of negligence.",
    "Personal injury and negligence claims require proof of harm."
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X)

print("=== Document Clustering (K=2) ===\n")
for i, (doc, label) in enumerate(zip(documents, kmeans.labels_)):
    print(f"Document {i+1} → Cluster {label + 1}")
    print(f"  \"{doc}\"\n")

print("Cluster Summary:")
for cluster_id in range(2):
    print(f"\n  Cluster {cluster_id + 1}:")
    for i, label in enumerate(kmeans.labels_):
        if label == cluster_id:
            print(f"    - Doc {i+1}: {documents[i]}")
```

## Exercise 2 — Cosine Similarity (TF-IDF)

```python
!pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

doc1 = "The tenant failed to pay rent on time, violating the lease agreement."
doc2 = "The lease was breached because the monthly rent payment was late."
doc3 = "The company merged with a competitor to dominate the market."

docs = [doc1, doc2, doc3]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

sim_1_2 = cosine_similarity(X[0], X[1])[0][0]
sim_1_3 = cosine_similarity(X[0], X[2])[0][0]

print("=== Cosine Similarity (TF-IDF) ===\n")
print(f"Doc1: \"{doc1}\"")
print(f"Doc2: \"{doc2}\"")
print(f"Doc3: \"{doc3}\"\n")
print(f"Similarity(Doc1, Doc2) = {sim_1_2:.4f}  → {'HIGH – similar topic' if sim_1_2 > 0.3 else 'MODERATE – some shared terms'}")
print(f"Similarity(Doc1, Doc3) = {sim_1_3:.4f}  → {'HIGH' if sim_1_3 > 0.3 else 'LOW – unrelated topics'}")
```

## Exercise 3 — SentenceTransformer Semantic Similarity

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer, util

doc1 = "The tenant failed to pay rent on time, violating the lease agreement."
doc2 = "The lease was breached because the monthly rent payment was late and not sent to the appropriate email."
doc3 = "The company merged with a competitor to dominate the market."

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode([doc1, doc2, doc3], convert_to_tensor=True)

sim_1_2 = util.cos_sim(embeddings[0], embeddings[1]).item()
sim_1_3 = util.cos_sim(embeddings[0], embeddings[2]).item()

print("=== Semantic Similarity (SentenceTransformer) ===\n")
print(f"Doc1: \"{doc1}\"")
print(f"Doc2: \"{doc2}\"")
print(f"Doc3: \"{doc3}\"\n")
print(f"Similarity(Doc1, Doc2) = {sim_1_2:.4f}  → {'SIMILAR' if sim_1_2 > 0.5 else 'NOT SIMILAR'}")
print(f"Similarity(Doc1, Doc3) = {sim_1_3:.4f}  → {'SIMILAR' if sim_1_3 > 0.5 else 'NOT SIMILAR'}")
print()
print("Interpretation:")
print(f"  Doc1 vs Doc2: {sim_1_2:.4f} — Both describe a lease/rent violation → HIGH semantic overlap")
print(f"  Doc1 vs Doc3: {sim_1_3:.4f} — Completely different topics → LOW semantic overlap")
```
